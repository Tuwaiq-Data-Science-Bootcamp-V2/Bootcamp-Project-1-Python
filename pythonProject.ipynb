```python
currentBoard = ['| 1 |' , '| 2 |' , '| 3 |' , '| 4 |' , '| 5 |' , '| 6 |' , '| 7 |' , '| 8 |' , '| 9 |'  ]
board = currentBoard.copy()
checkAvailble = [True , True , True ,True ,True ,True ,True ,True ,True ]
checkAvailble2 = checkAvailble.copy()
turn = True
currentMoves = {
    1 : '',
    2 : ''
}
moves = currentMoves.copy()


# function to draw the board
def drawBoard(board):
    for i in range( 0 , len(board) , 3):
        print(board[i] , board[i+1] , board[i+2])

```

```python
#function to place a player move in the board
def play(board , squareNum , checkAvailble , turn , moves):
    
    while True:
        while True:
            if squareNum > 8 or squareNum < 0:
                squareNum = int (input("The number must be from 1 - 9 , please choose another number: ")) - 1
            if squareNum <= 8 and squareNum >= 0:
                break
        squareNum1 = lambda a : a + 1 
        if checkAvailble[squareNum] and turn:
            board[squareNum] = '| X |'
            checkAvailble[squareNum] = False
            turn = False
            moves[1] +=str(squareNum1(squareNum))
            checkAvailble[squareNum] = False
            break
        elif checkAvailble[squareNum]:
            board[squareNum] = '| O |'
            checkAvailble[squareNum] = False
            turn = True
            moves[2] += str(squareNum1(squareNum))
            checkAvailble[squareNum] = False
            break
        else :
            squareNum = int (input("The square is unavalible , please choose another square: ")) - 1
    return turn
```
