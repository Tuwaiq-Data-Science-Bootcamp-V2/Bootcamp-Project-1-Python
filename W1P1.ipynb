```python
Accounts = {1234:{'Name':'Rahaf','Age': 22, 'Balance':4356}, 1444:{'Name':'Lamess','Age': 22, 'Balance':4000} }
i = 1

def AccountExists (ID):
    if ID in Accounts :
        return True
    else:
        return False
        
        
def NewAccount() :
    try:
        idA = int(input("Enter your id: "))
        if AccountExists(idA) == False :
                name = input("Enter your name: ")
                age = input("Enter your age: ")
                balance = float(input("Enter your balance: "))
                Accounts[idA] = {}
                Accounts[idA]["Name"] =name
                Accounts[idA]["Age"] =age
                Accounts[idA]["Balance"] =balance
                print("\nAccount has been successfully created")
           
        else:
            print("\nAccount already exits!\n")
            
                
            
    except ValueError:
        print("Invalid data type")

       
           
def DisplayAccount():
    try:
        ID = int(input("Enter your ID: "))
        if AccountExists (ID) == True:
            for pid in Accounts.items():
                if ID in pid :
                    print("\nAccount details")
                    print("Id: ",ID )
                    print("Name: ", Accounts[ID]["Name"])
                    print("Age: ", Accounts[ID]["Age"])
                    print("Balance: ", Accounts[ID]["Balance"])
                else:
                    print("\nAccount does not exist\n") 
            
    
    except ValueError:
        print("Invalid data type")
        
def NewBalance():
    try:
        ID = int(input("Enter your ID: "))
        if AccountExists (ID) == True:
            newB = float(input("Enter the amount: "))
            for pid in Accounts.items():
                if ID in pid :
                    bal= Accounts[ID]["Balance"]
                    neb = lambda a, b : a - b
                    Accounts[ID]["Balance"] = neb(bal, newB)
                    print("\nCurrent balance: \n", Accounts[ID]["Balance"])
                else: 
                    print("\nAccount does not exist\n")
        
    
    except ValueError:
        print("Invalid data type")


    
    
    
 

print("Welcome to tuwaiq bank\n")

while True :
    print("\n1.Make new account")
    print("2.Display account details")
    print("3.Deposit the amount")
    print("4.Exit\n")
    try:
        choice = int(input ("\nEnter your choice \n"))
        if choice == 1 :
            NewAccount()
        elif choice == 2:
            DisplayAccount()
        elif choice == 3:
            NewBalance()
        elif choice == 4:
            print("\n Goodbye\n")
            break
        else:
            print("Invalid choice!")
    except ValueError:
        print("Invalid data type")
        
    
        
    
```
