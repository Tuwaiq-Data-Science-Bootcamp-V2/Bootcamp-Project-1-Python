Order Menu

```python
Order_Menu = {1: ("Food", Food_Menu), 2: ("Drinks", Drinks_Menu),
              3: ("Desserts", Desserts_Menu), 4: "Back", 5: "Exit"}
```

```python
def show_menu(Menu):
    for i, option in enumerate(Menu):
        print(str(i+1) + '-', Menu[option])
```

```python
def edit_account(Client):
    while True:
        Client["Account"]["Email"] = input("Enter Your Email: ")
        Client["Account"]["Pass"] = input("Enter Your Pass: ")
        Client["Account"]["Name"] = input("Enter Your Name: ")
        Client["Account"]["Payment"]["Card Name"] = input(
            "Enter Your Card Name: ")
        Client["Account"]["Payment"]["Card Secret Number"] = int(
            input("Enter Your Card Name: "))
        print("Edited  Successfully")
        Login()
```

```python
def get_offer(Client):
    while True:
        show_menu(Offers_Menu)
        while True:
            Offer = int(input("Choose an Order:"))
            if Offer in Offers_Menu:
                if Offer == 6:
                    break
                Client["Orders"].append(Offers_Menu[Offer])
            else:
                print("I'm sorry, We don't serve that.")
            if is_order_complete():
                return output_order(Client)
            else:
                break
        return services_menu(Client)
```

```python
def get_order(Client):
    while True:
        show_menu(Order_Menu)
        while True:
            option = int(input("Choose an Option:"))
            if option == 1:
                show_menu(Food_Menu)
            elif option == 2:
                show_menu(Drinks_Menu)
            elif option == 3:
                show_menu(Desserts_Menu)
            elif option == 4:
                services_menu(Client)
            else:
                exit()
            while True:
                order = int(input("Choose your Order:"))
                if order in Order_Menu[option][1]:
                    if order == 6:
                        break
                    print(Client)
                    Client["Orders"].append(Order_Menu[option][1][order])
                else:
                    print("I'm sorry, We don't serve that.")
                    continue
                if is_order_complete():
                    return output_order(Client)
                else:
                    break
            break
```

```python
def services_menu(Client):
    show_menu(Services_Menu)
    while True:
        option = int(input("Choose an Option: "))
        if option == 1:
            get_order(Client)
        elif option == 2:
            get_offer(Client)
        elif option == 3:
            edit_account(Client)
        else:
            exit()
```

```python
def Login():
    while True:
        if not Clients:
            print("please signup first")
            return main_menu(Main_Menu)
        Email = input("Enter Your Email: ")
        for Client in Clients:
            if Client["Account"]["Email"] == Email:
                Password = input("Enter Your Password: ")
                if Client["Account"]["Pass"] == Password:
                    print("Successfully Logged in")
                    Client["Account"]["LoggedIn"] = True
                    break
        if Client["Account"]["LoggedIn"] == True:
            services_menu(Client)
        else:
            print("Wrong Credentials")
            main_menu(Main_Menu)
```

```python
def Signup():
    while True:
        Client["Account"]["Email"] = input("Enter Your Email: ")
        Client["Account"]["Pass"] = input("Enter Your Pass: ")
        Client["Account"]["Name"] = input("Enter Your Name: ")
        Client["Account"]["Payment"]["Card Name"] = input(
            "Enter Your Card Name: ")
        Client["Account"]["Payment"]["Card Secret Number"] = int(
            input("Enter Your Card Number: "))
        Clients.append(Client)
        print("Signup  Successfully")
        Login()
```

```python
def main_menu(Menu):
    print("Welcome to Alkhair Resturant")
    show_menu(Menu)
    while True:
        option = int(input("Choose an Option: "))
        if option == 1:
            Login()
        elif option == 2:
            Signup()
        else:
            exit()
```

```python
def is_order_complete():
    print("Anything else? yes/no")
    choice = input()
    if choice == "no":
        return True
    elif choice == "yes":
        return False
    else:
        raise Exception("invalid input")
```

```python
def output_order(Client):
    print("#################\n")
    print("Welcome to Alkhair Resturant")
    print("Your orders are: ")
    time = 0
    price = 0
    for i, order in enumerate(Client["Orders"]):
        print(
            f"{i+1}- {order['Name']}, Time:{str(order['Time'][0]) + order['Time'][1]}, Price:{order['Price']}")
        price += int(order['Price'].split(' ')[0])
        time += int(order['Time'][0])
    print("\n-----------------\n")
    print(
        f"Total waiting Time is: {time}\nTotal Price is: {price} syr\nCome back soon\n")
    exit()
```
