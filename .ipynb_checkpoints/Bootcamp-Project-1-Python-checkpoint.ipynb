```python
#AMC cinemas --online booking system
#global variables 
global_rating = 0
flag = True 
mv_dict = {"movie1":11, "movie2":10, "movie3":10, "movie4":10, "movie5":0} #movie dictionary contain movie name and number of seats
cart_dict = {"movie1":0, "movie2":0, "movie3":0, "movie4":0, "movie5":0} #cart dictionary
```

```python
#functions defention
dec_func = lambda a : a -1 #decrement function
inc_func = lambda a : a +1 #increment function
def view_mv(): #view movies and number of seats avaliable 
    for k, v in mv_dict.items():
        print('movie name: ', k , ' and there is ', v, ' seat avaliable')
def mv_assig(mv_name): #remove seat from movie dictionary and add it to the cart dictionary
    mv_dict[mv_name] = dec_func(mv_dict[mv_name])
    cart_dict[mv_name] = inc_func(cart_dict[mv_name])
    print("you booked ", cart_dict[mv_name], " (a) seat successfully and there is/are ", mv_dict[mv_name], " seat(s) remaining")            
def book_mv(): #book a movie from the avaliable list
    view_mv()
    book_flag = True
    while(book_flag):
        chosen_mv = input("please enter the movie name you would like to book ")
        if chosen_mv == "movie1":
            if (mv_dict["movie1"] >= 1):
                mv_assig("movie1")
                book_flag = False
            else:
                print("sorry, the movie seats are sold out.. try agin with another movie")
        elif chosen_mv == "movie2":
            if (mv_dict["movie2"] >= 1):
                mv_assig("movie2")
                book_flag = False
            else:
                print("sorry, the movie seats are sold out.. try agin with another movie")
        elif chosen_mv == "movie3":
            if (mv_dict["movie3"] >= 1):
                mv_assig("movie3")
                book_flag = False
            else:
                print("sorry, the movie seats are sold out.. try agin with another movie")
        elif chosen_mv == "movie4":
            if (mv_dict["movie4"] >= 1):
                mv_assig("movie4")
                book_flag = False
            else:
                print("sorry, the movie seats are sold out.. try agin with another movie")
        elif chosen_mv == "movie5":
            if (mv_dict["movie5"] >= 1):
                mv_assig("movie5")
                book_flag = False
            else:
                print("sorry, the movie seats are sold out.. try agin with another movie")
        else: 
            print("Error: you entered the wrong name, please try again")
def checkout(): #checkout function
    for k, v in cart_dict.items():
        print("movie name: ", k, " ", v, " seat(s)")
    checkout_flag = input("If you want to proceed to payement please enter (yes) otherwise the cart will be cleared")
    if checkout_flag == "yes" or checkout_flag == "Yes":
        print("Your reservation have been saved, thank you")
    else:
        for k, v in cart_dict.items():
            cart_dict[k] = 0
        for k, v in cart_dict.items():
            print("movie name: ", k, " ", v, " seat(s)")
        print("Your reservation have been canceled, thank you")
def rating(): #rate user experince
    rating_flag = True
    while(rating_flag):
        rating = int(input("Thank you.. enter a value from 1 to 5 to rate your experince "))
        if (rating >=1) and (rating <=5):
            global global_rating
            global_rating = global_rating+rating
            print("your feedback is important and the overall rating was ", global_rating)
            rating_flag = False
        else: 
            print("Error: only digit from 1 to 5, please try again")
def exit_q(): #to exit the program 
    inner_flag = True
    while (inner_flag):
        result= input("if you want to view previous menu enter 0, if you want to exit enter 00 ")
        if(result == "0"):
            inner_flag = False
            return True
        elif(result == "00"):
            rating()
            inner_flag = False
            return False
        else:
            print("Invalid choice please try again")
```

```python
#the beginning of the program
while(flag):
    print("please enter your choice number")
    print("enter 1 if you want to view the avaliable movies")
    print("enter 2 if you want to book a movie seat")
    print("enter 3 to proceed to checkout")
    print("enter 4 if you want to exit")
    user_choice = input()
    if(user_choice == "1"):
        view_mv()
        flag = exit_q()
    elif(user_choice == "2"):
        book_mv()
        flag = exit_q()
    elif(user_choice == "3"):
        checkout()
        flag = exit_q()
    elif(user_choice == "4"):
        rating()
        flag = False
    else:
        print("Invalid choice please try again")
```
